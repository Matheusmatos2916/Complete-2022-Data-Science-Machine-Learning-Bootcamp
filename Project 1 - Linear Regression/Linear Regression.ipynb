```python
import pandas as pd
from pandas import DataFrame as df
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression 

data = pd.read_csv("data.csv")

data.describe()
```

```python
a = df(data, columns = ['production_budget_usd'])
b = df(data, columns = ['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(a, b, alpha=0.3)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget ($)')
plt.ylabel('Worldwide Gross ($)')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000.0)
plt.show()
```

```python
regression = LinearRegression()
regression.fit(a,b)
```

## Slope coeficient

```python
regression.coef_ # theta_1
```

```python
#intercept
regression.intercept_
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(a, b, alpha=0.3)
plt.plot(a, regression.predict(a), color="red", linewidth=3)

plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget ($)')
plt.ylabel('Worldwide Gross ($)')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

```python
regression.score(a, b)
```
